```python
# 1 agglo
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import matplotlib.pyplot as plt

sample_df=pd.read_csv("spiral.txt")

training_points=sample_df[["xREAL","yREAL","class"]]
training_labels= sample_df["class"]

agglo= AgglomerativeClustering(n_clusters=5,linkage='single').fit(training_points)

plt.scatter(training_points["xREAL"],training_points["yREAL"],c=agglo.labels_,cmap='rainbow')

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
from sklearn.metrics.cluster import adjusted_rand_score


arc = adjusted_rand_score(training_labels, agglo.labels_)

print(arc)
```

```python
# 1 DBSCAN
from sklearn.cluster import DBSCAN
import pandas as pd
import matplotlib.pyplot as plt

sample_df = pd.read_csv("spiral.txt")

training_points = sample_df[["xREAL","yREAL","class"]]
training_labels = sample_df["class"]

dbscan = DBSCAN(eps=0.6, min_samples=2).fit(training_points)

plt.scatter(training_points["xREAL"], training_points["yREAL"], c=dbscan.labels_,
	cmap='rainbow')

plt.show()
```

```python
from sklearn.cluster import DBSCAN
import pandas as pd
from sklearn.metrics.cluster import adjusted_rand_score


arc = adjusted_rand_score(training_labels, dbscan.labels_)

print(arc)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.cluster import DBSCAN


sample_df = pd.read_csv("spiral.txt")

training_points=sample_df[["xREAL","yREAL","class"]]
training_labels=sample_df["class"]

max=0.0
a=0.1
b=1

for a in range(1,200,1):
  for b in range(1,20,1):
    dbscan=DBSCAN(eps=a*0.01, min_samples=b).fit(training_points)
    arc=adjusted_rand_score(training_labels, dbscan.labels_)
    if(max<arc):
      max=arc
      result_a=a
      result_b=b

print(result_a*0.01)
print(result_b)
print(max)

dbscan=DBSCAN(eps=result_a*0.01, min_samples=result_b).fit(training_points)
arc=adjusted_rand_score(training_labels, dbscan.labels_)
plt.scatter(training_points["xREAL"], training_points["yREAL"],c=dbscan.labels_,cmap='rainbow')
plt.show()
```
